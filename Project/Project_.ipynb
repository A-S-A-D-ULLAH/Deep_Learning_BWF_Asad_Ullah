## Combine csv files

```python
files=glob.glob('*.csv')
```

```python
combined = pd.DataFrame()

for file in files:
    data = pd.read_csv(file)
    combined = combined.append(data, ignore_index=True)
```

```python
combined.head()
```

```python
combined.dtypes
```

```python
combined.columns
```

```python
combined.head()
```

```python
combined[['male_population', 'female_population']] = combined['GenderPop'].str.split('_', expand=True)
```

```python

# Remove the 'M' and 'F' suffixes from the population columns
combined['male_population'] = combined['male_population'].str.replace('M', '').astype(int)
combined['female_population'] = combined['female_population'].str.replace('F', '0').astype(int)
```

```python
combined['female_population']=combined['female_population'].fillna(0)
```

```python
combined['female_population']
```

```python
combined.duplicated()
```

```python
combined.drop_duplicates()
```

```python
# Remove the '$' sign from the monetary column and convert it to numeric
combined['Income'] = combined['Income'].str.replace('$', '').astype(float)
```

```python
combined.drop_duplicates()
```

## Scatter

```python
income_by_state = combined.groupby('State')['Income'].mean()

proportion_women_by_state = combined.groupby('State')['female_population'].mean()

# Create a scatter plot of average income vs proportion of women
plt.scatter(proportion_women_by_state, income_by_state)

# Add axis labels and a title to the plot
plt.xlabel('Proportion of Women in the State')
plt.ylabel('Average Income by State')
plt.title('Scatter Plot of Average Income vs Proportion of Women')

# Show the plot
plt.show()
```

```python
combined['Hispanic'] = pd.to_numeric(combined['Hispanic'].fillna('0'),errors='coerce')


state_hispanic = combined.groupby('State')['Hispanic'].mean()

#proportion_women_by_state = combined.groupby('State')['female_population'].mean()

plt.hist(state_hispanic, bins=100)

plt.gca().set_xticklabels(['{:.0f}%'.format(x*10) for x in plt.gca().get_xticks()])
# Add axis labels and a title to the plot
#plt.xlabel('Proportion of Women in the State')
#plt.ylabel('Average Income by State')
plt.title('Scatter Plot of Hispanic vs states')

# Show the plot
plt.show()
```
